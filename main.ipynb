```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# Loading train.csv
uncleaned_df = pd.read_csv("data/train.csv")

# data cleaning
df = uncleaned_df.drop(columns=["Cabin", "PassengerId", "Name", "Ticket"])
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# encoding gender
df["Sex"] = df["Sex"].map({
    "male": 0,
    "female": 1
})
df = pd.get_dummies(df, columns=["Embarked"], dtype=int)

# preparing for training
x = df.drop(columns=["Survived"])
y = df["Survived"]

# preparing the testing data
test_data = pd.read_csv("data/test.csv")
cleaned_test_data = test_data.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
cleaned_test_data["Age"] = cleaned_test_data["Age"].fillna(cleaned_test_data["Age"].mean())
cleaned_test_data["Embarked"] = cleaned_test_data["Embarked"].fillna(cleaned_test_data["Embarked"].mode()[0])
cleaned_test_data["Fare"] = cleaned_test_data["Fare"].fillna(cleaned_test_data["Fare"].median())

cleaned_test_data["Sex"] = cleaned_test_data["Sex"].map({
    "male": 0,
    "female": 1
})
cleaned_test_data = pd.get_dummies(cleaned_test_data, columns=["Embarked"], dtype=int)

# training the model
model = LogisticRegression(max_iter=1000)
model.fit(x, y)
y_predicted = model.predict(cleaned_test_data)

submission_data = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": y_predicted
})

submission_data.to_csv("submission.csv", index=False)
```
